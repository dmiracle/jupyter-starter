```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
from scipy.optimize import fsolve

x = np.linspace(-1, 10, 100)
L = .15 * x
G = sigmoid(x)
def f(x):
    return .15*x-sigmoid(x)

root = fsolve(f, 7)

with plt.xkcd():
    fig = plt.figure()
    ax = fig.add_axes((.1, .2, .8, .7))
    ax.spines.right.set_color('none')
    ax.spines.top.set_color('none')
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlabel('increasing lib ownage ->')
    ax.plot(x, L, '-r', label="dead viewers")
    ax.plot(x, G, label="increase in viewers")
    ax.plot(x, G-L+.8, label="marginal live viewers")
    ax.legend()
    ax.set_title('Diminishing returns from owning libs \n vs. viewer deaths due to misinformation')
    ax.annotate(
        'Max marginal \n live viewers',
        xy=(1.5, 1.4), arrowprops=dict(arrowstyle='->'), xytext=(0, .8)
    )
    # ax.annotate(
    #     'Stay left of here',
    #     xy=(root, 1), arrowprops=dict(arrowstyle='->'), xytext=(6, 0)
    # )
```
